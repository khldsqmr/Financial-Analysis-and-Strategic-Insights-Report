### __Project 2: New Product Launch Financial Assessment__

__Scenario:__

You're evaluating the financial aspects of launching a new telecommunications product. You have cost estimates, revenue projections, and want to assess the return on investment.

__Tasks:__

*   Load the data containing information about cost of goods sold and revenue projections for the new product, for a two-year period.
*   Calculate profit and return on investment (ROI) for each month.
*   Visualize the revenue projection for the new product over time.
*   Print average monthly profit and average monthly ROI for the new product.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
# Load the pricing_data.csv
new_product_data_realistic = pd.read_csv("new_product_data_realistic.csv")

new_product_data_realistic.head()
```

```python
# Calculate total cost of goods sold
new_product_data_realistic['Cost_of_Goods_Sold'] = new_product_data_realistic['material_costs'] + new_product_data_realistic['labor_costs'] + new_product_data_realistic['other_costs']

# Calculate realistic revenue projection based on sales volume and pricing
new_product_data_realistic['Revenue_Projection'] = new_product_data_realistic['sales_volume'] * new_product_data_realistic['base_price'] * 1.25

new_product_data_realistic.head()
```

```python
# Calculate profit and return on investment
new_product_data_realistic['Profit'] = new_product_data_realistic['Revenue_Projection'] - new_product_data_realistic['Cost_of_Goods_Sold']
new_product_data_realistic['ROI'] = (new_product_data_realistic['Profit'] / new_product_data_realistic['Cost_of_Goods_Sold']) * 100

new_product_data_realistic.head()
```

```python
# Visualize revenue projection
plt.figure(figsize=(14, 4))

# Plotting the revenue projection
plt.plot(new_product_data_realistic['Month'], new_product_data_realistic['Revenue_Projection'], label='Revenue Projection', linestyle='--', marker='o', color='blue')

# Highlighting key points (e.g., milestones)
# You can replace these indices with the actual indices of significant events
significant_events = [3, 7, 10]
plt.scatter(new_product_data_realistic['Month'].iloc[significant_events], new_product_data_realistic['Revenue_Projection'].iloc[significant_events],
            color='red', marker='*', label='Milestones')

# Adding grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Adding labels and title
plt.title('Revenue Projection Over Time')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')

# Adding data labels in green color with a small font
for month, projection in zip(new_product_data_realistic['Month'], new_product_data_realistic['Revenue_Projection']):
    plt.text(month, projection, f'${projection:.0f}', fontsize=8, color='green', ha='center', va='bottom')

plt.legend()

# Show the plot
plt.show()
```

```python
# Assess profit and ROI metrics
average_profit_new_product_realistic = new_product_data_realistic['Profit'].mean()
average_roi_new_product_realistic = new_product_data_realistic['ROI'].mean()

print(f"Average Monthly Profit for New Product: ${average_profit_new_product_realistic:.2f}")
print(f"Average Monthly ROI for New Product: {average_roi_new_product_realistic:.2f}%")
```
