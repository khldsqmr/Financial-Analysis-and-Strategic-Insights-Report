### __Project 3: New Product Launch Financial Assessment__

__Scenario:__

You're assessing the financial viability of a potential business development opportunity in the telecommunications industry. You have data on costs, revenue projections, and associated risks.

__Tasks:__

*   Load the data containing information about costs, revenue projections, and a risk factor for a business development opportunity, for a three-year period.
*   Calculate profit and adjusted revenue considering the risk factor.
*   Visualize the revenue projection and adjusted revenue over the years.
*   Print the average annual profit for the business development opportunity.


```python
# Import the Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the pricing_data.csv
opportunity_data_realistic = pd.read_csv("opportunity_data_realistic.csv")

opportunity_data_realistic.head()
```

```python
# Calculate total costs
opportunity_data_realistic['costs'] = opportunity_data_realistic['marketing_costs'] + opportunity_data_realistic['research_costs'] + opportunity_data_realistic['development_costs']

# Calculate realistic revenue projection based on sales volume and pricing
opportunity_data_realistic['revenue_projection_opportunity'] = opportunity_data_realistic['sales_volume_opportunity'] * opportunity_data_realistic['base_price_opportunity']

# Calculate adjusted revenue considering the risk factor
opportunity_data_realistic['adjusted_revenue'] = opportunity_data_realistic['revenue_projection_opportunity'] * (1 - opportunity_data_realistic['risk_factor'])

opportunity_data_realistic.head()
```

```python
# Calculate profit and adjusted revenue considering the risk factor
opportunity_data_realistic['Profit'] = opportunity_data_realistic['adjusted_revenue'] - opportunity_data_realistic['costs']

opportunity_data_realistic.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming opportunity_data_realistic has columns 'Year', 'revenue_projection_opportunity', and 'adjusted_revenue' (replace with your actual column names)

# Set a professional color palette
colors = ['#1f78b4', '#33a02c']  # Blue and Green, you can adjust these colors based on your preference

# Visualize revenue projection and adjusted revenue using a bar plot
plt.figure(figsize=(14, 4))

bar_width = 0.35
bar_positions_opportunity = np.arange(len(opportunity_data_realistic['Year']))
bar_positions_adjusted = bar_positions_opportunity + bar_width

plt.bar(bar_positions_opportunity, opportunity_data_realistic['revenue_projection_opportunity'], width=bar_width, label='Revenue Projection', color=colors[0], edgecolor='black')
plt.bar(bar_positions_adjusted, opportunity_data_realistic['adjusted_revenue'], width=bar_width, label='Adjusted Revenue', color=colors[1], edgecolor='black')

# Adding data labels on top of the bars
for x, y1, y2 in zip(bar_positions_opportunity, opportunity_data_realistic['revenue_projection_opportunity'], opportunity_data_realistic['adjusted_revenue']):
    plt.text(x, y1, f'${y1:.2f}', ha='center', va='bottom', color=colors[0], fontsize=8)
    plt.text(x + bar_width, y2, f'${y2:.2f}', ha='center', va='bottom', color=colors[1], fontsize=8)

# Adding labels and title
plt.title('Revenue Projection vs. Adjusted Revenue Over Years')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.xticks(bar_positions_opportunity + bar_width / 2, opportunity_data_realistic['Year'])  # Centering x-ticks

# Adding legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Assess profit metrics
average_profit_opportunity_realistic = opportunity_data_realistic['Profit'].mean()

print(f"Average Annual Profit for Business Development Opportunity: ${average_profit_opportunity_realistic:.2f} million")
```
